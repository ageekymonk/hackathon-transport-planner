```python
user_travel = df.groupby(['MASKEDCIN']).filter(lambda x: len(x['MASKEDCIN']) >= 3).groupby(['MASKEDCIN'])
```

```python
from datetime import datetime
start_location = {}
# user_travel['JOURNEYSTARTTIME'] = datetime.strptime(user_travel['JOURNEYSTARTTIME'], '%d/%b/%y %I:%M:%S.000000000 %p')
for user, journeys in user_travel:
    journey_list = journeys[['ORIGINLAT', 'ORIGINLONG', 'ORIGINTSN', 'JOURNEYSTARTTIME']].values
    journey_list = map(lambda x: [x[0], x[1], x[2], datetime.strptime(x[3], '%d/%b/%y %I:%M:%S.000000000 %p')], journey_list)
    journey_list.sort(key=lambda x: x[3])
    one_journey = start_location.get((journey_list[0][0], journey_list[0][1]), 0) + 1
    start_location[(journey_list[0][0], journey_list[0][1])] = one_journey
    
print(start_location)
```

```python
import json
with open('/Users/ramz.sivagurunathan/hacks/hackathon/code/passenger_start.json', 'wb') as fp:
    output_arr = []
    for k,v in start_location.iteritems():
        output_arr.append({'lat' : k[0], 'long': k [1], 'count' : v})
        
    json.dump(output_arr, fp)
```
