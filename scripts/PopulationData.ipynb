```python
import csv
import json
tz_info = {}
with open('/Users/ramz.sivagurunathan/hacks/hackathon/repos/transport-analyser/data/empforecast.csv') as fp:
    reader = csv.reader(fp)
    for row in reader:
        if reader.line_num > 1:
            tz_info[row[0]] = {'employment' : row[1]}
print tz_info
```

```python
with open('/Users/ramz.sivagurunathan/hacks/hackathon/datasets/population/bts_population.csv') as fp:
    reader = csv.reader(fp)
    for row in reader:
        try:
            if reader.line_num > 1:
                tz_info[row[1]]['household'] = tz_info[row[1]].get('household',[])
                tz_info[row[1]]['household'].append(row[5])
                tz_info[row[1]]['population'] = tz_info[row[1]].get('population',[])
                tz_info[row[1]]['population'].append(row[2])
                tz_info[row[1]]['private'] = tz_info[row[1]].get('private',[])
                tz_info[row[1]]['private'].append(row[3])
                tz_info[row[1]]['nonprivate'] = tz_info[row[1]].get('nonprivate',[])
                tz_info[row[1]]['nonprivate'].append(row[4])
        except:
            pass

print tz_info
```

```python
with open('/Users/ramz.sivagurunathan/hacks/hackathon/repos/transport-analyser/data/emp_population.json','w') as fp:
    json.dump(tz_info, fp, indent=4)
```
