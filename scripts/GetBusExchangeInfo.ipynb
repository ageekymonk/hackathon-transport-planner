```python
import csv
interchanges_latlong = {}
with open('/Users/ramz.sivagurunathan/hacks/hackathon/datasets/routes/stops.txt') as fp:
    reader = csv.reader(fp)
    for row in reader:
        for interchange in interchanges:
            if row[2].find(interchange) != -1:
                interchanges_latlong[interchange] = interchanges_latlong.get(interchange, [])
                if (row[3],row[4]) not in interchanges_latlong[interchange]:
                    interchanges_latlong[interchange].append((row[3],row[4]))
                    interchangeInfo[interchange]['stops'].append({'lat' : row[3], 'long': row[4]})

print interchangeInfo
```

```python
interchange_stopids = {}
with open('/Users/ramz.sivagurunathan/hacks/hackathon/datasets/routes/stops.txt') as fp:
    reader = csv.reader(fp)
    i = 0
    for row in reader:
        lat, lon, platform = row[3], row[4], row[8]
        for ichange,lat_long_list in interchanges_latlong.iteritems():
            for lat_long in lat_long_list:
                if lat == lat_long[0] and lon == lat_long[1] and platform == "":
                    interchange_stopids[ichange] = interchange_stopids.get(ichange, [])
                    if row[0] not in interchange_stopids[ichange]:
                        interchange_stopids[ichange].append(row[0])
                        for idx, stop in enumerate(interchangeInfo[ichange]['stops']):
                            if lat == stop['lat'] and lon == stop['long']:
                                interchangeInfo[ichange]['stops'][idx]['id'] = row[0]
print interchangeInfo
```

```python
interchange_bus = {}
with open('/Users/ramz.sivagurunathan/hacks/hackathon/datasets/routes/stop_times.txt') as fp:
    reader = csv.reader(fp)
    for row in reader:
        trip_id, stop_id = row[0], row[3]
        if reader.line_num > 1:
            bus_no = trip_id.split('-')[1]
            for ichange, stop_id_list in interchange_stopids.iteritems():
                if stop_id in stop_id_list:
                    interchange_bus[ichange] = interchange_bus.get(ichange, [])
                    if bus_no not in interchange_bus[ichange]:
                        interchange_bus[ichange].append(bus_no)
                        interchangeInfo[ichange]['buses'].append(bus_no)
                    
print interchangeInfo
```
